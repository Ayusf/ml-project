```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the 'average' column exists
df['average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(25, 6))

# First subplot: Overall distribution by lunch type
sns.histplot(data=df, x='average', kde=True, hue='lunch', ax=axs[0])
axs[0].set_title("Average Score Distribution by Lunch Type (Overall)")

# Second subplot: Female students
sns.histplot(data=df[df.gender == 'female'], x='average', kde=True, hue='lunch', ax=axs[1])
axs[1].set_title("Average Score Distribution by Lunch Type (Female)")

# Third subplot: Male students
sns.histplot(data=df[df.gender == 'male'], x='average', kde=True, hue='lunch', ax=axs[2])
axs[2].set_title("Average Score Distribution by Lunch Type (Male)")

plt.show()
```
